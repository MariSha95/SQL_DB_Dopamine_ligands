```python
import pandas as pd
import sqlite3

# Чтение CSV-файла
df = pd.read_csv('enriched_dopamine_ec50.csv')

# Создание (или подключение к) базы данных SQLite
conn = sqlite3.connect('mydatabase.db')

# Запись данных в таблицу (если таблица не существует, она будет создана)
df.to_sql('dopamine_ligands', conn, if_exists='replace', index=False)

# Закрытие соединения
conn.close()
```

```python
conn = sqlite3.connect('mydatabase.db')
cursor = conn.cursor()
# Выбор определенных столбцов и фильтрация результатов, удовлетворяющих условию
cursor.execute("SELECT ChEMBL, EC50 FROM dopamine_ligands WHERE EC50< 1")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
# Фильтрация по диапазону значений молекулярной массы
cursor.execute("SELECT  CHEMBL, MW FROM dopamine_ligands WHERE MW BETWEEN 200 AND 220")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
# Фильтрация значений EC50 исходя из определенного типа рецептора
cursor.execute("SELECT  CHEMBL, EC50 FROM dopamine_ligands WHERE Target = 'Dopamine D3 receptor'")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
# Cреднее значениие молекулярной массы исследуемых веществ
cursor.execute("SELECT AVG(MW) AS MW  FROM dopamine_ligands")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
# Cреднее количество доноров водородной связи среди исследуемых лигандов
cursor.execute("SELECT AVG(H_Donors) AS H_Donors FROM dopamine_ligands")
result = cursor.fetchall()
for row in result:
    print(row)
```

```python
conn = sqlite3.connect('mydatabase.db')
query = "SELECT MW, EC50, LogP, TPSA, pEC50 FROM dopamine_ligands"
df = pd.read_sql_query(query, conn)

#Гистограмма распределения молекулярных масс
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
plt.figure(figsize=(10, 5))
sns.histplot(df['MW'], bins=50, kde=True)
plt.title('Гистограмма распределения молекулярных масс')
plt.xlabel('Молекулярная масса')
plt.ylabel('Частота')
plt.show()

#Гистограмма распределения значений EC50
plt.figure(figsize=(10, 5))
sns.histplot(df['EC50'], bins=50, kde=True)
plt.title('Гистограмма значений EC50')
plt.xlabel('EC50')
plt.ylabel('Частота')
plt.show()
```

```python
#Тепловая карта корреляций
corr = df.corr()

# Создаем тепловую карту
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)

# Настраиваем заголовок и показываем график
plt.title('Тепловая карта корреляций')
plt.show()
```
